```python
!pip install pyinputplus
import pyinputplus as pyip

# Display a menu for choosing between temperature conversion and wind chill calculation.
def menu():
  print("Weather Calculator - A W4 Project \n 1. From Celcius to Fahrenheit \n 2. From Fahrenheit to Celcius \n 3. Wind Chill Calculator \n 4. Exit")
  choice = pyip.inputChoice(['1', '2', '3', '4'], prompt = "Enter your choice (1-4): ")
  return choice

# Get the necessary inputs (temperature in Fahrenheit, and for wind chill, also wind speed).
def get_number(prompt):
  while True:
    try:
      return float(input(prompt))
    except ValueError:
      print("Please enter a valid number.")
      continue

def calculate_windchill(temp, wind_speed):
  result = 35.74 + 0.6215 * temp - 35.75 * wind_speed**0.16 + 0.4275 * temp * wind_speed**0.16
  return result

# Perform the chosen calculation using the formulas provided.
def calculate_temp(operation, num):
  if operation == '1':
    result = num * 9/5 + 32
  elif operation == '2':
    result = (num - 32) * 5/9
  else:
    return "Invalid choice"
  return result

#Main Script
choice = menu()
if choice == '4':
  print("Thank you, see you again")
elif choice == '3':
  temp = get_number("Enter temperature in Fahrenheit: ")
  wind_speed = get_number("Enter wind speed: ")
  result = calculate_windchill(temp, wind_speed)
  print("Result:", result)
else:
  num = get_number("Enter temperature: ")
  result = calculate_temp(choice, num)
  print("Result:", result)
```
